```python
import matplotlib.pyplot as plt
parks = pd.read_csv('https://raw.githubusercontent.com/ernbilen/Data200_Fall23/main/data/national_parks.csv')
```

```python
parks
```

```python
import numpy as np
np.sqrt(parks.shape[0])
```

```python
plt.hist(parks.GrandCanyon,edgecolor='black',facecolor='springgreen',bins=7)

plt.ylim(0,30)

plt.xlabel('# of visitors in million people')
plt.ylabel('count')

plt.savefig('/users/bilene/desktop/grand-canyon.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
# scatterplot
plt.scatter(parks.Year,parks.GrandCanyon)

plt.xlabel('year')
plt.ylabel('# of visitors')

plt.show()
```

```python
# line plot
plt.plot(parks.Year,parks.GrandCanyon)

plt.xlabel('year')
plt.ylabel('# of visitors')

plt.show()
```

```python
# line plot
plt.plot(parks.Year,parks.GrandCanyon,label='GrandCanyon')
plt.plot(parks.Year,parks.BryceCanyon,label='BryceCanyon')
plt.plot(parks.Year,parks.Badlands,label='Badlands')

plt.xlabel('year')
plt.ylabel('# of visitors')

plt.legend()

plt.show()
```

```python
# practice time:
# work on the countries data
# make two histograms on the same plot for lifeExpectancy, one for where year is 2007 and one for 1952
# use different colors for the histograms
# show legend
# set alpha=.75 for 2007, and alpha=.5 for 1952
```

```python
plt.hist(countries[countries.year==1952].lifeExpectancy, alpha=.5, facecolor="blue", \
         edgecolor='black',bins=12, label="1952")
plt.hist(countries[countries.year==2007].lifeExpectancy, alpha=.75, facecolor="purple", \
         edgecolor='black',bins=12, label="2007")
plt.xlabel("life expectancy")
plt.ylabel("count")
plt.legend()
plt.show()
```

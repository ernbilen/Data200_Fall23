```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ernbilen/Data200_Fall23/main/data/indicators-main.csv')
```

```python
# set the index of df to be the code column
df.index = df.code
```

```python
# return countries whose gdp is less than 200, but only return their
# country name and population
# regular pandas syntax:
df[df.gdp<200][['country','pop']]

# .loc method:
df.loc[df.gdp<200,['country','pop']]
```

```python
# appending dfs

import pandas as pd

data1 = {'country': ['Canada', 'China', 'India', 'Russia', 'United States', 'Vietnam'], 
		'pop': [36.26, 1378.66, 1324.17, 144.34, 323.13, 94.59], 
		'gdp': [1535.77, 11199.15, 2263.79, 1283.16, 18624.47, 205.28],
		'life': [82.30, 76.25, 68.56, 71.59, 78.69, 76.25], 
		'cell': [30.75, 1364.93, 1127.81, 229.13, 395.88, 120.60]}

data2 = {'country': ['Germany','United Kingdom'],
	'pop': [82.66, 66.06], 
	'gdp': [3693.20, 2637.87],
	'life': [81.16, 81.16], 
	'cell': [80.75, 92.23]}
ind1 = pd.DataFrame(data1)
ind2 = pd.DataFrame(data2)
```

```python
ind1
```

```python
ind2
```

```python
# you could append them together with append()
ind1.append(ind2)
```

```python
# this is the way to go
pd.concat([ind1,ind2]).reset_index(drop=True)
# axis=0 is the default option
```

```python
# what if you wanna keep an id for where your came from?
# create a new col 'group' equals 1 if the data is from ind1
# and equals 2 if the data is from ind2, then concat them together
ind1['group'] = 1
ind2['group'] = 2

pd.concat([ind1,ind2],axis=0).reset_index(drop=True)
```

```python
# how to combine two datasets that came from two sources
# e.g., df1 on Dickinson students (includes their home state)
# and df2 on sociodemographics on states in the U.S.

# toy example:
data1 = {'country': ['Canada', 'China', 'India', 'Russia', 'United States', 'Vietnam'], 
		'pop': [36.26, 1378.66, 1324.17, 144.34, 323.13, 94.59], 
		'gdp': [1535.77, 11199.15, 2263.79, 1283.16, 18624.47, 205.28],
		'life': [82.30, 76.25, 68.56, 71.59, 78.69, 76.25], 
		'cell': [30.75, 1364.93, 1127.81, 229.13, 395.88, 120.60]}
ind1 = pd.DataFrame(data1)

imports_exportsDoL = {
'country': ['India','China','United States'],
'imports': [2241.66, 392.23, 1601.76],
'exports': [1504.57, 266.16, 2280.54]}
imports_exports = pd.DataFrame(imports_exportsDoL)
```

```python
ind1
```

```python
imports_exports
```

```python
# join() function to combine datasets if the key variable is set as
# index, otherwise use merge()
ind1.merge(imports_exports,on='country',how='left') # default is 'inner'
```

```python
# merging topnames with indicators
indicators = pd.read_csv('https://raw.githubusercontent.com/ernbilen/Data200_Fall23/main/data/indicators-main.csv')
topnames = pd.read_csv('https://raw.githubusercontent.com/ernbilen/Data200_Fall23/main/data/topnames.csv')
```

```python
topnames.head()
```

```python
indicators[indicators.country=='United States']\
                        .merge(topnames, on='year', how='inner')
```

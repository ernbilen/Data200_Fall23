```python
%%sql
-- Return countries (from the countries table) with land>10000 and the country
-- should be either high income or upper middle income

SELECT * FROM countries
WHERE land > 10000 AND (income="High income" OR income="Upper middle income")

```

```python
%%sql
-- return the number of unique baby names in the topnames table

SELECT COUNT(DISTINCT name) FROM topnames
```

```python
%%sql
-- return a descriptive statistics table which contains the following
-- mean pop, mean gdp, mean life, mean cell, count of non-missing obs
-- from the indicators table

SELECT ROUND(AVG(pop),2) AS avg_pop, ROUND(AVG(gdp),2) AS avg_gdp,
ROUND(AVG(life),2) AS avg_life, ROUND(AVG(cell),2) AS avg_cell,
COUNT(*) AS n 
FROM indicators
```

```python
%%sql
--filter the data to country observations where gdp is 
--greater than the average gdp in the gdp
SELECT * FROM indicators
WHERE gdp>(SELECT AVG(gdp) FROM indicators)
```

```python
%%sql
--return the total land size of the earth excluding ANTARTICA
--using the countries table
SELECT SUM(land) AS total_land FROM countries
```

```python
%%sql
-- returns the unique values in region column
SELECT DISTINCT(region) FROM countries
```

```python
%%sql
--return the total land size by region
SELECT region, SUM(land) AS tot_land 
FROM countries
GROUP BY region
ORDER BY tot_land DESC
```

```python
%%sql
--return the total land size by region
-- note that because we are using group by we have to
-- use HAVING instead of WHERE
SELECT region, SUM(land) AS tot_land 
FROM countries
GROUP BY region
HAVING tot_land>20000000
ORDER BY tot_land DESC
```

```python
%%sql
-- Return the top 10 most popular baby names that ever existed
-- in the U.S. history using the topnames table
SELECT name, SUM(count) AS tot_count FROM topnames
GROUP BY name
ORDER BY tot_count DESC
LIMIT 10
```

```python
%%sql
-- return the average year by baby name, add this info to the
-- table above
SELECT name, SUM(count) AS tot_count, ROUND(AVG(year),0) AS avg_year 
FROM topnames
GROUP BY name
ORDER BY tot_count DESC
LIMIT 10
```

```python
%%sql
-- Return the top 10 most popular baby names that existed
-- in the U.S. after the year 2000 using the topnames table
-- include the count of obs
SELECT name, SUM(count) AS tot_count, COUNT(*) 
FROM topnames
WHERE year>2000
GROUP BY name
--HAVING year>2000
ORDER BY tot_count DESC
```

```python
%%sql
-- Suppose a student has registered for four classes with classid
-- fields given as 21014, 21088, 21256, 21444. Return the names
-- of the instructors who teach these courses. Include their
-- instructorid, first and last name, their departmentid, and division
SELECT instructorid,instructorfirst,instructorlast,departmentid, division FROM instructors
INNER JOIN instructor_class USING(instructorid)
INNER JOIN departments USING(departmentid)
WHERE classid IN (21014,21088,21256,21444)
```

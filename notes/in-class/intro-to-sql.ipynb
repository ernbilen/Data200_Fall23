```python
%%sql
-- Choosing or "selecting" columns
SELECT year AS Year,name FROM topnames
LIMIT 5
```

```python
%%sql
-- Choosing or "filtering" rows
SELECT year AS Year,name FROM topnames
WHERE Year>=2000 AND count>30000
```

```python
%%sql
-- Return rows where baby name is not John
-- In pandas: name!='John'
-- In SQL: name <> 'John'
SELECT * FROM topnames
WHERE name <> 'John'
```

```python
%%sql
-- Filtering on categorical data?
-- Return high income countries
SELECT * FROM countries
WHERE income='High income'
```

```python
%%sql
-- Filter based on intervals
-- Return rows where 'life' is between 65 and 75
SELECT * FROM indicators
WHERE life BETWEEN 65 AND 75
```

```python
%%sql
-- Filter using a list of elements
-- Return countries with 'code' CHN, IND, FRA, USA
SELECT * FROM indicators
WHERE code IN ('CHN','IND','FRA','USA')
```
